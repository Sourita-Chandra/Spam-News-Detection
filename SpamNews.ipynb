# NLP

## Preprocessing the data

```python
import nltk

nltk.download('stopwords')

from nltk.corpus import  stopwords
from nltk.stem import WordNetLemmatizer
```

```python

ps = WordNetLemmatizer()
```

```python

stopwords = stopwords.words('english')
```

```python

import re
```

```python

def clean_msg(row):
    row = row.lower()
    row = re.sub('[^a-zA-Z]',' ',row)
    
    token = row.split()
    news = [ps.lemmatize(word) for word in token if not word in stopwords]
    
    cleaned_news = ' '.join(news)
    return cleaned_news
```

```python

import nltk
nltk.download('wordnet')

data['text'] = data['text'].apply(lambda x: clean_msg(x))
```

```python
data['text']     
```

```python
## convert all textual data into numbers
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

vectorizer = TfidfVectorizer(max_features = 50000, lowercase = False, ngram_range = (1,2))
```

```python

x = data.iloc[:35000, 0]
y = data.iloc[:35000, 1]
```

```python
x.head()
```

```python

from sklearn.model_selection import train_test_split
```

```python

train_data, test_data, train_label, test_label = train_test_split(x, y, test_size = 0.2)
```

```python
train_data
```

```python

vtrain_data = vectorizer.fit_transform(train_data)
```

```python
vtrain_data
```

```python

vtrain_data = vtrain_data.toarray()
```

```python
vtrain_data
```

```python

vtest_data = vectorizer.fit_transform(test_data)

vtest_data = vtest_data.toarray()
```

```python

training_data = pd.DataFrame(vtrain_data, columns= vectorizer.get_feature_names_out())
testing_data = pd.DataFrame(vtest_data, columns= vectorizer.get_feature_names_out())
```

```python
training_data
```

## Modeling 

```python

from sklearn.naive_bayes import MultinomialNB
```

```python

clf = MultinomialNB()
```

```python

clf.fit(training_data, train_label)
```

```python

y_pred = clf.predict(testing_data)
```

```python

from sklearn.metrics import accuracy_score
```

```python

accuracy_score(test_label, y_pred)
```
